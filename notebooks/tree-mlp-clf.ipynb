```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import seaborn as sns
import scipy
from scipy.stats import pearsonr

import sklearn
from sklearn import datasets, linear_model
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

#read file
all_df=pd.read_csv('NPHA-doctor-visits.csv', index_col=False)
all_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

X = all_df.drop('Number of Doctors Visited', axis=1)
le=LabelEncoder()
all_df['Number of Doctors Visited']=le.fit_transform(all_df['Number of Doctors Visited'])
all_df.head()
y=all_df['Number of Doctors Visited']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import cross_val_score

n_fold=5
depth_range = range(3, 16)
best_depth = 0
best_score = 0

# evaluate each depth
for depth in depth_range:
    
    tree_clf = DecisionTreeClassifier(max_depth=depth)
    scores = cross_val_score(tree_clf, X, y, cv=n_fold)
    mean_score = 100*np.mean(scores)
    print(f"depth:{depth},score:{mean_score:.3f}%")
    
    #record the best score
    if mean_score > best_score:
        best_score = mean_score
        best_depth = depth

print(f"Best depth：{best_depth}，Best score：{best_score:.3f}%")
```

```python
from sklearn.neural_network import MLPClassifier

mlp_clf = MLPClassifier(random_state=1,max_iter=1500)
scores = cross_val_score(mlp_clf, X, y, cv=5)
print(scores)
avg = (100 * np.mean(scores), 100 * np.std(scores) / np.sqrt(scores.shape[0]))
print("Average score and standard deviation: (%.2f +- %.3f)%%" % avg)
```
