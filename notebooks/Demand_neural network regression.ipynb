```python
X = all_df.drop('Target (Total orders)', axis=1)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Xs = scaler.fit_transform(X)

y = all_df['Target (Total orders)']
```

```python
from sklearn.model_selection import train_test_split

Xs_train, Xs_test, y_train, y_test = train_test_split(Xs, y, test_size=0.3, random_state=1)
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

kf = KFold(n_splits=5, shuffle=True, random_state=1)
```

```python
mlp_regressor = MLPRegressor(hidden_layer_sizes=(50,), max_iter=50000, random_state=1)
mse_scores = cross_val_score(mlp_regressor, Xs, y, cv=kf, scoring='neg_mean_squared_error')
mse_scores = -mse_scores

print(f'Cross-validated Mean Squared Error scores: {mse_scores}')
print(f'Average MSE: {np.mean(mse_scores):.2f}')
print(f'Standard Deviation of MSE: {np.std(mse_scores):.2f}')
```

```python
mlp_regressor = MLPRegressor(hidden_layer_sizes=(100,), max_iter=50000, random_state=1)

mse_scores = cross_val_score(mlp_regressor, Xs, y, cv=kf, scoring='neg_mean_squared_error')
mse_scores = -mse_scores

print(f'Cross-validated Mean Squared Error scores: {mse_scores}')
print(f'Average MSE: {np.mean(mse_scores):.2f}')
print(f'Standard Deviation of MSE: {np.std(mse_scores):.2f}')
```
