```python
X = all_df.drop('Number of Doctors Visited', axis=1)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Xs = scaler.fit_transform(X)

y = all_df['Number of Doctors Visited']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
Xs_train, Xs_test, y_train, y_test = train_test_split(Xs, y, test_size=0.3, random_state=1, stratify=y)

from sklearn.linear_model import LogisticRegression
param_grid = {'C': [0.01, 0.1, 1, 10, 100]}

log_reg = LogisticRegression(random_state=1)
grid_search = GridSearchCV(log_reg, param_grid, cv=5, scoring='accuracy')
grid_search.fit(Xs_train, y_train)

print(f"Best parameter C: {grid_search.best_params_['C']}")

best_log_reg = grid_search.best_estimator_
classifier_score = best_log_reg.score(Xs_test, y_test)
print(f'The classifier accuracy score with best parameter C is {classifier_score:.2f}')
```
